# Welcome to VU Evolving Landscapes collection!

VU Evolving Landscapes contains a selected number of notebooks providing an introduction to landscape evolution modeling. Part of this material is based on content written and maintained by the [Community Surface Dynamics Modeling System][CSDMS] (CSDMS). For more extensive notebooks and background, see the [CSDMS Ivy collection][csdms-ivy]. 

## Goal 

The goal of these lessons is to become familiar to landscape evolution modeling software. Through this collection of tutorials, students will be introduced to landscape evolution modeling and will be able to investigate how the interactions between tectonics and geomorphic processes are expressed in topography. 

## Lessons 

1. Introduction to landscape evolution modeling
2. Hillslopes
3. Rivers
